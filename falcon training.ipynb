```python
list_train = []
for i in range(10000):
    store = {}
    store['instruction'] = "Summarize the following conversation"
    store['input'] = dataset['train']['dialogue'][i]
    store['output'] = dataset['train']['summary'][i]
    list_train.append(store)
```

```python
import json
with open('./summary_data.json', 'w') as file:
    json.dump(list_train, file)
```
